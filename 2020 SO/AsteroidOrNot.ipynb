```
import numpy as np
import pylab as plt
```

```
import pandas as pd
import io
import requests
url="https://ssd.jpl.nasa.gov/dat/ELEMENTS.UNNUM"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')),delim_whitespace=True,skiprows=3)
```

```
cols = ['Designation', 'CODE','Epoch', 'a', 'e', 'i', 'w', 'Node', 'M', 'H', 'G', 'Ref','Jiji']
c.columns = cols
```

```
def NEO(a):
  if a>1: return 1 - a**-1
  else: return a**-1 - 1
ass = np.linspace(0.5,5,100)


fig,ax = plt.subplots(figsize=(8,8))
plt.plot(c['a'],c['e'],'k.',ms=0.3,alpha=1)

neos = [NEO(a) for a in ass]
plt.plot(ass,neos,'b-',lw=2)
ax.fill_between(ass,neos,1,alpha=0.1,color='b')
ax.annotate("NEAs",(0.9,0.4),size=30,color='b',weight="bold")

ax.annotate("2020 SO",(1.1,0.03),size=12,color='r',weight="bold")
plt.plot(1.034,0.036,'r*',ms=20)

plt.xlim([0.6,3])
plt.ylim([0,0.6])

plt.title("Elementos Orbitales de 2020 SO \n comparado con otros Asteroides",size=20)
plt.xlabel("Semieje Mayor [AU]",size=15)
plt.ylabel("Excentricidad",size=15)
```
