```python
import numpy as np
import pylab as plt
from scipy.special import sph_harm,eval_genlaguerre
```

```python
fac = np.math.factorial
sh = sph_harm #m,l,theta,phi
lg = eval_genlaguerre #n,l

def phi(r,th,phi,N,L,M):
    a0 = 1
    rho = 2*r/(N*a0)
    A = (2 / N / a0)**1.5 * (fac(N-L-1)/(2*N*fac(N+L)))**.5
    B = np.exp(-rho/2)
    C = lg(N-L-1,2*L+1,rho).real
    D = sh(M,L,phi,th).real
    
    return abs(A*B*C*D*rho**L)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(plt.imread("Hydrogen_Density_Plots.png"))
```

```python
Nr = 50
Nth = 50

N = 3
L = 1
M = 0

size = 10
fig = plt.figure(figsize=(size,size))

def subplot(subnum,N,L,M,R):
    rs = np.linspace(0,R,Nr)
    ths = np.linspace(0,2*np.pi,Nth)
    tt,rr = np.meshgrid(ths,rs)
    
    
    ax = fig.add_subplot(subnum, projection='polar')
    values = phi(rr,tt,0,N,L,M)
    ax.contourf(ths,rs,values,100,cmap='magma')
    ax.grid(False)
    ax.set_yticks([])
    ax.set_xticks([])
    ax.set_title("n={:.0f} ; l={:.0f} ; m={:.0f}".format(N,L,M))
    
subplot(441,1,0,0,1)

subplot(445,2,0,0,10)
subplot(446,2,1,-1,10)
subplot(447,2,1,0,10)
subplot(448,2,1,1,10)
```
